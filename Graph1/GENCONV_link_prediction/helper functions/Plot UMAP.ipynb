```python
import umap
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
## upload any embedding file you want to plot ##
##-----------------------------------------------

## this is the lncRNA embeddings from GENCNV model for the combined db

embedding_path = "Graph2-embeddings-GENCONV/lncRNA_embedding.pt"
embeddings = torch.load(embedding_path)
#print(embeddings.shape)
#print(embeddings[:2])  # Show first two embeddings
```

```python
def plot_umap(embeddings, n_neighbors=15, min_dist=0.1, random_state=42):
    """
    Plot UMAP projection of input embeddings.

    Args:
        embeddings (numpy.ndarray or torch.Tensor): The input node embeddings of shape [N, D].
        n_neighbors (int): UMAP parameter controlling local connectivity.
        min_dist (float): UMAP parameter controlling clustering tightness.
        random_state (int): Random seed for reproducibility.
    """
    # Convert to numpy if tensor
    if isinstance(embeddings, torch.Tensor):
        embeddings = embeddings.detach().cpu().numpy()

    reducer = umap.UMAP(n_neighbors=n_neighbors, min_dist=min_dist, random_state=random_state)
    umap_proj = reducer.fit_transform(embeddings)

    # Plot
    plt.figure(figsize=(8, 6))
    plt.scatter(umap_proj[:, 0], umap_proj[:, 1], s=10, alpha=0.8)
    plt.title("UMAP Projection of Embeddings")
    plt.xlabel("UMAP-1")
    plt.ylabel("UMAP-2")
    plt.grid(False)
    plt.show()
```

```python
plot_umap(embeddings)
```
